```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
import sklearn
```

```python
pic = plt.imread('images/PCAM8.png')
print(pic.shape)
temp = pic.shape[2]
plt.imshow(pic)
```

```python
pic_n = pic.reshape(pic.shape[0]*pic.shape[1], pic.shape[2])
pic_n.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, n_init = 10, random_state=0).fit(pic_n)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python
cluster_pic = pic2show.reshape(pic.shape[0], pic.shape[1], pic.shape[2])
plt.imshow(cluster_pic)
```

```python
newpic = (cluster_pic * 255).astype(np.uint8)
h,w,bpp = np.shape(newpic)
for py in range(0,h):
    for px in range(0,w):
        if(newpic[py][px][0] > 140):            
            newpic[py][px][0]=26
            newpic[py][px][1]=43
            newpic[py][px][2]=235


plt.imshow(newpic)
```
